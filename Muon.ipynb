### Background

Muon's are fundamental particles, that act as an unstable heavy electron.

Cosmic rays, made up of about 98% protons and 2% electrons (plus some other unimportant stuff) collide with the nuclei of air molecules, which in turn create a number of particles, one of which is the pion

$
\pi^+ = \mu^+ \nu_{\mu}
$

$
\pi^{-} = \mu^{-} \overline{\nu_{\mu}}
$

This is specific to spontaneous decay via the weka force. Because of this the muon only interacts with matter through weak and electromagnetic forces. This causes it to decay into an electron plus a neutrino and antineutrino while losing its kinetic energy. We can detect this electron with the scintillator as an electric pulse which has some value for the time first measured plus time to stop measuring.

    We also have negligible input from the neutrinos and antineutrinos

    Here is an example of muon detection from a cosmic ray

Because some of the particles decay, we only have a certain amount of muons that survive down to sea-level, this value is approximately:
$
\Phi = 1 min^{-1}cm^{-2}
$

With $KE = 4 GeV$

The interesting aspect of muons is that because they are produced roughly at 15km above sea level, their decay time would not normally allow them to reach sea level. However, because of time dilation, they have an extended lifetime in our frame due to how fast they are travelling.

We can find the decay time for muons using differential equations. Because we can describe the probability of a muon decaying over a small time interval dt we have some constant out from for the decay rate, resulting in:  
$
dN = -N(t)\lambda dt
$  
Where \lambda is the decay rate, and N(t) is the number of surviving muons at time t. This is a common differential equation known to be of exponential form:
$
N(t) = N_0 e^{-\lambda t}
$  
However, we only detect a probability of muons decaying inside of our scintillator. Therefore we can redefine N(t) in terms of D(t) which is the decay probability inside our scintillator, which is what we are measuring.

Our muon lifetime would then be $\tau = \frac{1}{\lambda}$

The value that we measure will be the amount of time that a muon takes to decay inside the scintillator. Because a muon can start decay outside of the scintillator, we need to be sure of how we define what we are measuring. The exponential graph should still describe the muon decay time because we are measuring a large number of particles, so the number that starts decay outside the scintillator will be greatest and decay so on and so forth. Therefore, the graph we receive will just be the same function, with some standard error. As the manual says: "An exponential function always "looks the same" in the sense that whether you examine it at early times or late times, its e-folding time is the same"

Here are some properties of our scintillator:  


As you would note with the pi interaction with matter, we can also produce some negative muons. These muons bind with the scintillator's carbon and hydrogen nuclei in the same way that electrons do. It can occupy an already filled atomic orbit that has electrons, which can in turb then interact with protons. Because we measure "charge species" with the scintillator we average over both the positive and negative charge species of the muons, this will cause us to have a lifetime slightly less than that of the free space value:
$
\tau_\mu = 2.19703 \pm 0.00004 \mu s
$

The free space value can be considered the expected lifetime from positive muons in the plastic scintillator. We can also take the the lifetime of negative mons to be:  
$
\tau_c = 2.043 \pm 0.003
$ [Reiter, 1960]  

With the equation for the average observed decay rate, we can calculate the the momentum range of muons entering the scintillator:  
$
<\lambda>  = \frac{N^+ \lambda^+ + N^- \lambda^-}{N^+ + N^-} = \frac{\rho \lambda^+ + \lambda^-}{1 + \rho}
$  
$
\tau_{obs} = (1 + \rho) \frac{\tau^- \tau^+}{\tau^+ + \rho \tau^-}
$  
We can rearrange this to find $\rho$:  
$
\rho = -\frac{\tau^+}{\tau^-} \left(\frac{\tau^- - \tau_{obs}}{\tau^+ - \tau_{obs}} \right)
$

We can't necessarily move the scintillator in our lab experiment, so we will only be calculating the momentum at the altitude of our lab at UCSC; this is roughly 230 m above sea level.

#### Time Dilation Effect

For the time dilation effect, we are supposed to measure the stopping rate at two different altitudes, however, we do not have the ability to do that in our lab. Instead we will use their recommendation at the end of this section at doing a lower altitude and predicting the value at a higher altitude.

For our scintillator, the average stopping rate is around 160 MeV. The stopping rate is found with the total # of muon decays recorded by the instrument in some time interval. We can find this value with the second input from our data collection which is the local time as measured by the computer.

Energy is an increasing function with height, so our altitude should yield higher results than what would be expected at sea level. We can use this to estimate the value of energy at higher altitudes.

The average energy loss rate of muons travelling close to the speed of light is around 2 MeV/g/cm^2 (Get a source for this). A more accurate version is given with the Bethe-Blcoh equation:  
$
\frac{dE}{dx} = 0.3071 \left( \frac{MeV}{g cm^{-2}} \right) \frac{Z}{A} \rho \frac{1}{\beta^2} z^2 \left( ln \frac{2mc^2}{I} \beta^2 \gamma^2 - \beta^2 \right)
$  
N is the number of electrons in the stopping medium per cm^3, e is the electronic charge, z is the atomic number, Z is the atomic number, A is the atomic weight (of the stopping medium, beta is the velocity of th eprojectile in units respective to the speed of light, gamma is the Lorentz factor, I is the mean excitation energy of the stopping medium atoms roughly equal to AZ.

We can roughly estimate the energy loss of a muon as:  
$
\Delta E = 2 MeV g^{-1} cm^{-2} * H * \rho_{air}
$  
With p being the density of air: $\rho_{air} = \rho_0 e^{-h/h_0}$ for p_0 = 1030 g/cm^2 and h_0 = 8400 m

For the particles rest frame we have:  
$
t^{'} = \frac{mc}{\rho C_0} \int_{\gamma_1}^{\gamma_2} \frac{1}{\sqrt{\gamma^2 - 1}} \,d\gamma
$  
Here C_0 = 2 MeV g^-1 cm^2 (get clarity on this cm^2 whether its ^- or ^+), gamma 2 = 1.5 (we choose this because we want this to represent the average muon to get stopped halfway through the scintillator). The lower limit of integration is gamma1 = E1/mc^2 for E1 = E2 + delE (this is our value).

We want to find the ratio of muon stopping rates for the detector at two positions:  
$
R = e^{-t^{'} / \tau}
$  
Tau here is the muon proper lifetime that is specified in the manual. 

From here we want to compute the ratio between raw stopping rates, which we use to calculate the transit time in the muon's frame. We compare this to the theoretical stopping rate ratio, R_0 = R_{raw}/R, which we shall use as the correction factor.

We will use the correction factor and divide our lower elevation stopping rate by it and then account for energy loss and predict the upper elevation. We will compare this value against a measurement. For example:  
$
R_{\mu} = R_0 R
$

Here we expect the function for muon decay to take the form:
$
N(t) = N_0 e^{-\lambda t}

```python
def monoExp(x, m, t1, b):
    return m * np.exp([-t1 * item for item in x]) + b
```

```python
p0 = (4482, .413, 0) # I did a very bad method of this, just guessing and checking values to get hte best R^2 value that I could
params, cv = scipy.optimize.curve_fit(monoExp, y_arr, t, p0)
m, t1, b = params
b = AVG_VAL -19 # b was initially some intercept value, that should actually just be the average of the last set of values, according the the R^2 fit, this is currently good
# The sample rate is very important, here's the value found from the input data
sampleRate = 1110000 # Hz # (number of measurements from  171200a### to 171200(a+1)###) * 1000 (multiplying by 1000 to go from nanosecond to microsecond
tauSec = (1 / t1) / sampleRate

# determine quality of the fit
squaredDiffs = np.square(y_arr - monoExp(t, m, t1, b))
squaredDiffsFromMean = np.square(y_arr - np.mean(y_arr))
rSquared = 1 - np.sum(squaredDiffs) / np.sum(squaredDiffsFromMean)
print(f"RÂ² = {rSquared}")
```

```python
# plot the results
roundSec = round(tauSec*1e6, 3)
fig2, ax2 = plt.subplots()

markers, caps, bars = ax2.errorbar(t, y_arr, yerr = y_mn, c='black', ls='none', capsize=5, elinewidth=1, markeredgewidth=1)
[bar.set_alpha(0.5) for bar in bars]
[cap.set_alpha(0.5) for cap in caps]

ax2.plot(t, y_arr, '.', label="data")
ax2.plot(t, monoExp(t, m, t1, b), '--', label=f'Y = {m} * e^(-(1/{roundSec}) x) + {b}')

ax2.set_xlabel('Muon Decay Time (Âµs)')
ax2.set_ylabel('Frequency (Events/Bin)')
ax2.set_title("Fitted Exponential Curve")

# inspect the parameters
print(f"Y = {m} * e^(-{t} * x) + {b}")
print(f"Tau = {tauSec * 1e6} Â± 0.1 Âµs")

plt.grid(alpha=.4,linestyle=':')
ax2.set_xlim([t[0]-0.5, t[-1]+1.5])
ax2.set_ylim([0, 7500])
plt.legend(fontsize=10)
fig2.savefig('fit.eps', format='eps')
plt.show()
```
